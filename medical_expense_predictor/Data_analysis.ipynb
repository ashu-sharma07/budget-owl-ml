```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
df
```

```python
df.describe()
```

```python
# Categorical Variables

for i in df:
    if df[i].dtype==object:
        print(i,end=" ")
        print("=>   Unique Values:-", end=" ")
        print(df[i].unique(),end="\t")
        print(f"Dtype of {i}:-",end=" ")
        print(df[i].dtype)
        
#Numerical files

for i in df:
    if df[i].dtype!=object:
        print(i,end=" ")
        print(f"Dtype of {i}:-",end=" ")
        print(df[i].dtype)
```

```python
#Na values
# make a list of the variables that contain missing values
vars_with_na = [var for var in df.columns if df[var].isnull().sum() > 0]

# determine percentage of missing values (expressed as decimals)
# and display the result ordered by % of missin data

df[vars_with_na].isnull().mean().sort_values(ascending=False)
```

No NA values are present. Pheww!!!!

```python
# i just printed all the unique values with their share just to get an idea about how the data is distributed:)
for var in df:
    print(df[var].value_counts(normalize=True)*100)
```

```python
for var in df:
    (df[var].hist(bins=50))
    plt.ylabel('No of Students')
    plt.xlabel('Monthly Expense')
    plt.show()
```

We observe that the data distribution is not skewed(in case of numerical categories).So no data needs to be transformed.
